## 1. The data

In this lab we will work on data from [ObsEra](http://www.ipgp.fr/en/obsera/obsera-observatory-of-water-and-erosion-in-the-caribbean). As an introduction to our purpose, we propose you to test some tools and commands on this dataset. The data is saved under the `./data` directory, in which there are the `data/CHEM` and `data/HYDRO` directories. We are only focusing on the first one. 

```python
import glob
import pandas as pd

# Path to chemical data
filepaths_chem = glob.glob("./data/CHEM/*.csv")
filepath_chem = filepaths_chem[0]

# Read in chemical data
data_chem = pd.read_csv(filepath_chem, sep=';')
```

## 2. Inspect with `pandas`

### 2.1. Basic inspection

There are many ways you can use `pandas` to inspect your data. The first one is to use the `head` method, which displays the first 5 rows of your dataset. You can also use the `tail` method to display the last 5 rows. Other methods such as the `describe` method can be used to get a quick description of your data, as well as the `info` method. The `info` is very useful to see the type of each fields especially when they are not numerical but categorical which of course may orient the choice of your machine-learning models. The `describe` focus on the numerical data giving very basic statistics for each fields.

```python
data_chem.head()
```

```python
data_chem.tail()
```

```python
data_chem.info()
```

```python
data_chem.describe()
```

The `pandas` library also allows to select data based on fields. For example, if you want to select the `Conductivity` field, you can use the following command. You can see the `pandas.DataFrame` objects as a `numpy.array` with named columns. You can also select multiple fields by passing a list of fields to the `[]` operator. Note that a single column of a `pandas.DataFrame` object 

```python
data_chem["Conductivity"].describe()
```

### 2.2. Basic plots of the data 

The `pandas` library also takes benefits from the `matplotlib` library to plot the data. For example, you can plot the `Conductivity` field using the `plot` method. You can also plot multiple fields by passing a list of fields to the `plot` method. 

```python
data_chem["Conductivity"].plot(style=".", grid=True, ylabel="Conductivity")
```

You can also use the `hist` method to plot the data distribution for each field.

```python
data_chem.hist(figsize=(10, 10));
```

### 2.3. More addvanced plotting

Several plotting functions are available in the `pandas` library. For example, you can use the `scatter_matrix` function to plot the scatter matrix of the data, which will allow you to get insights about the correlation between the different fields. 

> Note that we use the semi-colon `;` at the end of the last line to avoid the output of the `scatter_matrix` function to be displayed. This is because the `scatter_matrix` function returns a `numpy.array` of `matplotlib.axes.Axes` objects, which are displayed by default when the cell is executed.

Take the time to understand and conclude about the correlation between the different fields. 

```python
from pandas.plotting import scatter_matrix

# Select attributes to plot (you can play around with this)
attributes = ["Conductivity", "Suspended Load", "Twater (°C)"] 

# Plot
scatter_matrix(data_chem[attributes]);
```
